```python
# Import necessary libraries  
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.cluster import AffinityPropagation  

# Set random seed for reproducibility  
np.random.seed(42)  

# Generate synthetic data  
n_samples = 300  
# Generate random 2D points around three centers  
X1 = np.random.randn(n_samples, 2) + np.array([5, 5])  
X2 = np.random.randn(n_samples, 2) + np.array([-5, -5])  
X3 = np.random.randn(n_samples, 2) + np.array([5, -5])  

# Combine the groups into one dataset  
X = np.concatenate([X1, X2, X3], axis=0)  

# Visualize the synthetic data  
plt.figure(figsize=(8, 6))  
plt.scatter(X[:, 0], X[:, 1], color='blue', s=10, alpha=0.5)  
plt.title('Synthetic Data Points')  
plt.xlabel('Feature 1')  
plt.ylabel('Feature 2')  
plt.grid()  
plt.show()  

# Fit the Affinity Propagation model  
preference_value = -2  # Adjusting the preference value affects the number of clusters  
af = AffinityPropagation(preference=preference_value)  
af.fit(X)  

# Get cluster centers and labels  
cluster_centers_indices = af.cluster_centers_indices_  
labels = af.labels_  

# Visualize the clustering results  
plt.figure(figsize=(8, 6))  
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='rainbow', marker='o', alpha=0.7)  

# Marking cluster centers  
plt.scatter(X[cluster_centers_indices, 0],   
            X[cluster_centers_indices, 1],   
            s=200, c='black', marker='X', label='Cluster Centers')  # Exemplar points  

plt.title('Affinity Propagation Clustering Results')  
plt.xlabel('Feature 1')  
plt.ylabel('Feature 2')  
plt.legend()  
plt.grid()  
plt.colorbar(scatter, label='Cluster Label')  
plt.show()  

# Print cluster information  
print("Number of clusters:", len(cluster_centers_indices))  
print("Cluster Centers Indices:", cluster_centers_indices)  
print("Labels for each point:", labels)
```
